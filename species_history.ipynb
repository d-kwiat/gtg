```python
import coalestr as cs
```

```python
my_species = cs.species()

''' Default settings that can be modified using these keyword arguments

founder_duration = 1000, 
founder_N = 10,
founder_Q = 3,
founder_X = 0.2,

R0 = 1.0005,
expansion_Q = 3,
expansion_X = 0.2,

plateau_duration = 40000,
plateau_N = 14660,
plateau_Q = 3,
plateau_X = 0.2 '''

history = my_species.t_his

print("History of my species goes back", str(history), "generations")

observe = [x for x in range(0,history,5000)]
my_species.get_coalescent(observe, show = False)
my_species.get_diversity(show = False)
my_species.plot_observations(metrics = ("Nh", "snp_het",))
```

```python
plateau = 40000 # duration of plateau phase

my_species = cs.species(
    founder_duration = 1000, 
    founder_N = 10,
    founder_Q = 1,
    founder_X = 0,
    R0 = 1.0005,
    expansion_Q = 1,
    expansion_X = 0,
    plateau_duration = plateau,
    plateau_N = 18000,
    plateau_Q = 1,
    plateau_X = 0)

history = my_species.t_his

print("History of my species goes back", str(history), "generations")

observe = [x for x in range(0,plateau,5000)] + [x for x in range(plateau, history, 1000)]
my_species.get_coalescent(observe, show = False)
my_species.get_diversity(show = False)
my_species.plot_observations(metrics = ("Nh", "snp_het",))
```

```python
plateau = 40000 # duration of plateau phase

my_species = cs.species(
    founder_duration = 1000, 
    founder_N = 10,
    founder_Q = 10,
    founder_X = 1,
    R0 = 1.0005,
    expansion_Q = 10,
    expansion_X = 1,
    plateau_duration = plateau,
    plateau_N = 3200,
    plateau_Q = 10,
    plateau_X = 1)

history = my_species.t_his

print("History of my species goes back", str(history), "generations")

observe = [x for x in range(0,plateau,5000)] + [x for x in range(plateau, history, 1000)]
my_species.get_coalescent(observe, show = False)
my_species.get_diversity(show = False)
my_species.plot_observations(metrics = ("Nh", "snp_het",))
```
